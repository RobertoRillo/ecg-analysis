```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
testdata = pd.read_csv('mitbih_test.csv', header=None)
traindata = pd.read_csv('mitbih_train.csv', header=None)
print(testdata.head())
testdata.columns
```

```python
testdata
```

```python
# Asignar números como nombres de columnas después de cargar los datos
testdata.columns = range(testdata.shape[1])
traindata.columns = range(traindata.shape[1])

# Verifica que los encabezados se han asignado correctamente
print(testdata.head())
print(testdata.columns)
```

```python
X = traindata.iloc[:, :-1]  # Características
y = traindata.iloc[:, -1]   # Etiquetas
```

```python
X_tezt = testdata.iloc[:, :-1]  # Características
y_tezt = testdata.iloc[:, -1]   # Etiquetas
```

```python
print(traindata.iloc[:, -1].value_counts())
```

```python
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import classification_report
import numpy as np

# Definir el modelo de Random Forest
model = RandomForestClassifier(random_state=42)

# Definir el espacio de hiperparámetros
param_dist = {
    'n_estimators': [100, 200, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Validación cruzada estratificada (5 pliegues)
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Aplicar RandomizedSearchCV
random_search = RandomizedSearchCV(
    model, 
    param_distributions=param_dist,
    n_iter=20,  # Número de combinaciones a probar
    cv=kf,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# Para almacenar las predicciones y etiquetas reales
cv_y_test = []
cv_y_pred = []

# Realizar validación cruzada
for train_index, test_index in kf.split(X, y):
    # Separar datos de entrenamiento y prueba
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Aplicar undersampling para reducir el tamaño de la clase mayoritaria
    undersample = RandomUnderSampler(sampling_strategy='auto')
    X_train_under, y_train_under = undersample.fit_resample(X_train, y_train)

    # Aplicar SMOTE para balancear las clases
    smote = SMOTE()
    X_train_res, y_train_res = smote.fit_resample(X_train_under, y_train_under)

    # Aplicar RandomizedSearchCV para entrenar y optimizar hiperparámetros
    random_search.fit(X_train_res, y_train_res)

    # Predecir sobre el conjunto de prueba con el mejor modelo
    y_pred = random_search.best_estimator_.predict(X_test)

    # Almacenar los resultados para análisis final
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)

# Unir todas las predicciones y etiquetas verdaderas
cv_y_test = np.concatenate(cv_y_test)
cv_y_pred = np.concatenate(cv_y_pred)

# Reportar los resultados de la validación cruzada
print(classification_report(cv_y_test, cv_y_pred))

# Imprimir el mejor conjunto de hiperparámetros
print("Mejores hiperparámetros:", random_search.best_params_)
```

Excelente! Has encontrado los mejores hiperparámetros con RandomizedSearchCV. Hacer una validación cruzada con los hiperparámetros óptimos para ver si el desempeño mejora consistentemente.

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.metrics import classification_report
import numpy as np

# Definir el modelo de Random Forest con los mejores hiperparámetros
model = BalancedRandomForestClassifier(
    n_estimators=500,
    min_samples_split=5,
    min_samples_leaf=2,
    max_depth=20,
    bootstrap=False,
    class_weight='balanced', 
    random_state=42
)

# Validación cruzada estratificada (5 pliegues)
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Para almacenar las predicciones y etiquetas reales
cv_y_test = []
cv_y_pred = []

# Realizar validación cruzada
for train_index, test_index in kf.split(X, y):
    # Separar datos de entrenamiento y prueba
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Aplicar undersampling para reducir el tamaño de la clase mayoritaria
    undersample = RandomUnderSampler(sampling_strategy={0.0: 20000})
    X_train_under, y_train_under = undersample.fit_resample(X_train, y_train)

    # Aplicar SMOTE para balancear las clases
    smote = SMOTE()
    X_train_res, y_train_res = smote.fit_resample(X_train_under, y_train_under)

    # Entrenar el modelo en este pliegue con los hiperparámetros óptimos
    model.fit(X_train_res, y_train_res)

    # Predecir sobre el conjunto de prueba
    y_pred = model.predict(X_test)

    # Almacenar los resultados para análisis final
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)

# Unir todas las predicciones y etiquetas verdaderas
cv_y_test = np.concatenate(cv_y_test)
cv_y_pred = np.concatenate(cv_y_pred)

# Reportar los resultados de la validación cruzada
print(classification_report(cv_y_test, cv_y_pred))
```

MATRIZ DE CONFUSION CON DATOS TRAIN!!!

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Generar la matriz de confusión
cm = confusion_matrix(cv_y_test, cv_y_pred)

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[0, 1, 2, 3, 4], yticklabels=[0, 1, 2, 3, 4])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
